### 2.1: Tratar Tipos de Dados

```python
def convert_date_format(df, column_name, row_range=None):
    """
    Converte as datas no formato 'YYYY-MM-DD' para 'M/D/YYYY' e as transforma em datetime.
    
    Parâmetros:
        df (pd.DataFrame): DataFrame contendo a coluna de datas.
        column_name (str): Nome da coluna com as datas.
        row_range (tuple, opcional): Intervalo de linhas a serem convertidas (início, fim).
    
    Retorna:
        pd.DataFrame: DataFrame com a coluna de datas convertida.
    """
    if row_range:
        start, end = row_range
        rows = range(start, end)
    else:
        rows = df.index  # Aplica a conversão em todas as linhas
    
    for index in rows:
        old_date = df.at[index, column_name]
        parts = old_date.split('-')
        year, month, day = parts[0], parts[1], parts[2]
        new_date = f"{int(month)}/{int(day)}/{year}"
        df.at[index, column_name] = new_date
    
    df[column_name] = pd.to_datetime(df[column_name])
    return df
```

### 2.2: Criar Dataframes "r_fighters" e "b_fighters"

```python
def create_fighters_df(ufc_df, prefix):
    # Seleciona as colunas relevantes com base no prefixo
    selected_columns = [
        f"{prefix}fighter", "weight_class", f"{prefix}wins", f"{prefix}losses", f"{prefix}draw", f"{prefix}age",
        f"{prefix}win_by_Submission", f"{prefix}win_by_KO/TKO", f"{prefix}Stance", f"{prefix}total_title_bouts",
        f"{prefix}total_rounds_fought", f"{prefix}longest_win_streak", f"{prefix}win_by_Decision_Majority",
        f"{prefix}win_by_Decision_Split", f"{prefix}win_by_Decision_Unanimous",
        f"{prefix}win_by_TKO_Doctor_Stoppage", "date"
    ]
    
    # Cria um DataFrame com as colunas selecionadas
    fighters_df_raw = ufc_df[selected_columns].copy()
    
    # Dicionário de renomeação para remover o prefixo
    column_rename_dict = {col: col.replace(prefix, "") for col in selected_columns if col.startswith(prefix)}
    fighters_df_raw.rename(columns=column_rename_dict, inplace=True)
    
    # Ordena os dados por lutador, categoria de peso e data (decrescente)
    fighters_df_raw.sort_values(by=['fighter', 'weight_class', 'date'], ascending=[True, True, False], inplace=True)
    
    # Remove duplicatas mantendo apenas o registro mais recente por lutador e categoria de peso
    fighters_df = fighters_df_raw.drop_duplicates(subset=['fighter', 'weight_class'])
    
    return fighters_df
```

```python
# Criando os DataFrames para lutadores do corner vermelho e azul
r_fighters_df = create_fighters_df(ufc_df, "R_")
b_fighters_df = create_fighters_df(ufc_df, "B_")

r_fighters_df
```

### 2.3: Criar Dataframe "ufc_fighters"

```python
def process_fighter_stats(r_fighters_df, b_fighters_df):
    """
    Processa os DataFrames de lutadores do canto vermelho e azul, unifica os dados
    e calcula as colunas adicionais como o diferencial de vitórias e derrotas.
    
    Parâmetros:
        r_fighters_df (pd.DataFrame): DataFrame com informações dos lutadores do canto vermelho.
        b_fighters_df (pd.DataFrame): DataFrame com informações dos lutadores do canto azul.
    
    Retorna:
        pd.DataFrame: DataFrame concatenado com as colunas processadas.
    """
    # Concatenar os DataFrames ao longo das linhas
    ufc_fighters_df_raw = pd.concat([r_fighters_df, b_fighters_df], ignore_index=True)

    # Criar coluna 'win_by_Decision' somando os diferentes tipos de decisão
    ufc_fighters_df_raw['win_by_Decision'] = (
        ufc_fighters_df_raw['win_by_Decision_Majority'] +
        ufc_fighters_df_raw['win_by_Decision_Split'] +
        ufc_fighters_df_raw['win_by_Decision_Unanimous']
    )

    # Criar coluna 'win-loss differential'
    ufc_fighters_df_raw['win-loss differential'] = ufc_fighters_df_raw['wins'] - ufc_fighters_df_raw['losses']

    # Ordenar o DataFrame por 'fighter', 'weight_class' e 'date' (descendente para manter os dados mais recentes)
    ufc_fighters_df_raw.sort_values(by=['fighter', 'weight_class', 'date'], ascending=[True, True, False], inplace=True)

    return ufc_fighters_df_raw
```

```python
def aggregate_fighter_stats(ufc_fighters_df_raw):
    """
    Agrega as estatísticas dos lutadores, combinando categorias de peso e mantendo as estatísticas mais recentes.
    
    Parâmetros:
        ufc_fighters_df_raw (pd.DataFrame): DataFrame processado com informações do lutador.
    
    Retorna:
        pd.DataFrame: DataFrame com as estatísticas agregadas.
    """
    # Manter apenas a entrada mais recente de cada lutador por categoria de peso
    ufc_fighters_df = ufc_fighters_df_raw.drop_duplicates(subset=['fighter', 'weight_class'])

    # Definir colunas para agregação
    stat_columns = [
        'wins', 'draw', 'losses', 'age', 'win_by_Submission', 'win_by_KO/TKO', 'Stance',
        'total_title_bouts', 'total_rounds_fought', 'longest_win_streak', 'win_by_Decision_Majority',
        'win_by_Decision_Split', 'win_by_Decision_Unanimous', 'win_by_TKO_Doctor_Stoppage',
        'win_by_Decision', 'win-loss differential'
    ]

    # Agrupar por lutador, combinando categorias de peso e mantendo a data mais recente
    ufc_fighters_df = ufc_fighters_df.groupby('fighter', as_index=False).agg({
        'weight_class': '/'.join,  # Combina todas as categorias de peso do lutador
        'date': 'max',  # Mantém a data mais recente
        'age': 'max',  # Mantém a idade mais alta registrada
        **{col: 'sum' for col in stat_columns if col not in ['fighter', 'date', 'age', 'Stance']},
    })

    # Renomear a coluna para evitar problemas com barras no nome
    ufc_fighters_df = ufc_fighters_df.rename(columns={'win_by_KO/TKO': 'win_by_KO_TKO'})

    return ufc_fighters_df
```

```python
def process_fighters_data(r_fighters_df, b_fighters_df):
    """
    Processa os dados dos lutadores do UFC, unificando e agregando as estatísticas.
    
    Parâmetros:
        r_fighters_df (pd.DataFrame): DataFrame com informações dos lutadores do canto vermelho.
        b_fighters_df (pd.DataFrame): DataFrame com informações dos lutadores do canto azul.
    
    Retorna:
        pd.DataFrame: DataFrame processado e consolidado.
    """
    ufc_fighters_df_raw = process_fighter_stats(r_fighters_df, b_fighters_df)
    ufc_fighters_df = aggregate_fighter_stats(ufc_fighters_df_raw)

    return ufc_fighters_df
```

# Etapa 3: Análise de Apostas

```python
# Função para calcular o lucro ou prejuízo de uma aposta
def calcular_resultado(odds, winner, corner_escolhido, aposta=100):
    if winner == corner_escolhido:
        # Corner escolhido venceu: calcular lucro
        if odds > 0:
            return (odds / 100) * aposta
        else:
            return (100 / abs(odds)) * aposta
    else:
        # Corner escolhido perdeu: perder o valor da aposta
        return -aposta
```

```python
# Aplicar a função para simular apostas no corner vermelho
ufc_df['resultado_red'] = ufc_df.apply(
    lambda row: calcular_resultado(row['R_odds'], row['Winner'], 'Red'), axis=1
)

# Calcular o resultado total
resultado_total_red = ufc_df['resultado_red'].sum()

print(f"Resultado total apostando no corner vermelho: R${resultado_total_red:.2f}")
```

```python
# Aplicar a função para simular apostas no corner azul
ufc_df['resultado_blue'] = ufc_df.apply(
    lambda row: calcular_resultado(row['B_odds'], row['Winner'], 'Blue'), axis=1
)

# Calcular o resultado total
resultado_total_blue = ufc_df['resultado_blue'].sum()

print(f"Resultado total apostando no corner azul: R${resultado_total_blue:.2f}")
```

```python
# Extrair o ano da coluna 'date' (que já está no formato datetime)
ufc_df['year'] = ufc_df['date'].dt.year

# Agrupar os resultados por ano
resultados_por_ano_red = ufc_df.groupby('year')['resultado_red'].sum().cumsum()
resultados_por_ano_blue = ufc_df.groupby('year')['resultado_blue'].sum().cumsum()

# Criar o gráfico
plt.figure(figsize=(12, 6))
plt.plot(resultados_por_ano_red.index, resultados_por_ano_red, label='Corner Vermelho (Red)', marker='o', color='red')
plt.plot(resultados_por_ano_blue.index, resultados_por_ano_blue, label='Corner Azul (Blue)', marker='o', color='blue')

# Adicionar os valores no último ponto de cada linha
plt.text(resultados_por_ano_red.index[-1], resultados_por_ano_red.iloc[-1], f'{resultados_por_ano_red.iloc[-1]:.2f}', color='red', ha='center', va='bottom')
plt.text(resultados_por_ano_blue.index[-1], resultados_por_ano_blue.iloc[-1], f'{resultados_por_ano_blue.iloc[-1]:.2f}', color='blue', ha='center', va='bottom')

# Adicionar detalhes ao gráfico
plt.title('Resultado Acumulado das Apostas por Ano', fontsize=16)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Resultado Acumulado (R$)', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```
